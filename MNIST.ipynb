Work On MNIST Dataset BY using tensorflow and keras


```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np

# Import necessary libraries for building and evaluating the neural network
```

```python
(x_train,y_train),(x_test, y_test) = keras.datasets.mnist.load_data();

# Load the MNIST dataset, splitting it into training and testing sets
```

```python
x_train.shape

# Display the shape of the training data (number of samples, image height, image width)
```

```python
y_test.shape

# Display the shape of the test labels (number of test samples)
```

```python
x_train[0]

# Display the pixel values of the first image in the training set
```

```python
y_train[0]

# Display the label of the first image in the training set
```

```python
x_train_flattened = x_train.reshape(len(x_train),28*28)
x_test_flattened = x_test.reshape(len(x_test),28*28)

# Flatten the training and testing images from 28x28 to a 1D array of 784 pixels
```

```python
x_train_flattened[0]

# Display the flattened pixel values of the first training image
```

```python
np.max(x_train_flattened)

# Find the maximum pixel value in the flattened training data
```

```python
x_train_flattened = x_train_flattened/255
x_test_flattened = x_test_flattened/255

# Normalize the pixel values to be between 0 and 1 by dividing by the maximum value (255)
```

```python
x_train_flattened[0]

# Display the normalized flattened pixel values of the first training image
```

```python
model = keras.Sequential([
        keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
    ])

# Define the first simple neural network model
```

```python
model.compile(optimizer='adam', loss = "sparse_categorical_crossentropy", metrics = ['accuracy'])
model.fit(x_train_flattened, y_train , epochs=5)

# Compile the first model with an Adam optimizer and sparse categorical crossentropy loss, then train it for 5 epochs
```

```python
model.evaluate(x_test_flattened, y_test)

# Evaluate the first model's performance on the flattened test data
```

```python
plt.matshow(x_test[9])

# Display the 10th image from the original test set
```

```python
y_predicted = model.predict(x_test_flattened)
y_predicted[9]

# Make predictions on the flattened test data using the first model and display the prediction for the 10th image
```

```python
np.argmax(y_predicted[9])

# Find the index of the highest probability in the prediction for the 10th image (this is the predicted digit)
```

```python
y_predicted_label =[ np.argmax(i) for i in y_predicted]

# Get the predicted digit (the index with the highest probability) for each image in the test set
```

```python
y_predicted_label[:5]

# Display the first 5 predicted labels
```

```python
y_test[:5]

# Display the first 5 true labels from the test set
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_label)
cm

# Calculate the confusion matrix to evaluate the first model's performance
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')

# Visualize the confusion matrix as a heatmap
```

```python
# Adding hidden layer
# This section defines and trains a second model with a hidden layer for improved performance.
```

```python
md= keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10, activation = 'sigmoid')
])

# Define the second neural network model with a hidden layer
```

```python
md.compile(optimizer = 'adam', loss = "sparse_categorical_crossentropy", metrics = ['accuracy'])
md.fit(x_train_flattened, y_train, epochs = 5)

# Compile the second model and train it on the flattened training data
```

```python
md.evaluate(x_test_flattened, y_test)

# Evaluate the second model's performance on the flattened test data
```

```python
y_test_pred = md.predict(x_test_flattened)
y_test_pred[0]

# Make predictions on the flattened test data using the second model and display the prediction for the first image
```

```python
y_test[0]

# Display the true label of the first image in the test set
```

```python
np.argmax(y_test_pred[0])

# Find the index of the highest probability in the prediction for the first image (this is the predicted digit)
```

```python
y_test_pred_label = [np.argmax(i) for i in y_test_pred]

# Get the predicted digit for each image in the test set using the second model
```

```python
y_test_pred_label[:5]

# Display the first 5 predicted labels from the second model
```

```python
y_test[:5]

# Display the first 5 true labels from the test set
```

```python
# cm1 = tf.math.confusion_matrix(labels=y_test,prediction = y_test_pred_label)
cm1 = tf.math.confusion_matrix(labels=y_test, predictions=y_test_pred_label)

# Calculate the confusion matrix for the second model
```

```python
cm1

# Display the confusion matrix for the second model
```

```python
plt.figure(figsize=(10,7))
sn.heatmap(cm1,annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')

# Visualize the confusion matrix for the second model as a heatmap
```

```python
# Flateen layer by keras
# This section demonstrates using a Flatten layer within the Keras Sequential model.
```

```python
x_train1 = x_train/255
x_test1 = x_test/255

# Normalize the original training and testing data by dividing by 255
```

```python
md1= keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,activation='relu'),
    keras.layers.Dense(10, activation = 'sigmoid')
])

# Define a third model using a Flatten layer as the first layer
```

```python
# md1.compile(optimizer='adam', loss ="sparse_categorical_crossentropy" , metrics =['accuracy'])
# md1.fit(x_train_flattened, y_train, epochs = 5)
md1.compile(optimizer = 'adam', loss = "sparse_categorical_crossentropy", metrics = ['accuracy'])
md1.fit(x_train1,y_train, epochs = 5)

# Compile the third model and train it on the normalized original training data
```

```python
md1.evaluate(x_test1,y_test)

# Evaluate the third model's performance on the normalized original test data
```

```python
y_pred = md1.predict(x_test1)
y_pred[1]

# Make predictions on the normalized original test data using the third model and display the prediction for the second image
```

```python
np.argmax(y_pred[1])

# Find the index of the highest probability in the prediction for the second image (this is the predicted digit)
```

```python
y_test[1]

# Display the true label of the second image in the test set
```

```python
y_pred_label = [np.argmax(i) for i in y_pred]
y_pred_label[:5]

# Get the predicted digit for each image in the test set using the third model and display the first 5 predicted labels
```

```python
y_test[:5]

# Display the first 5 true labels from the test set
```

```python
cm2 = tf.math.confusion_matrix(labels=y_test, predictions=y_pred_label)
cm2

# Calculate the confusion matrix for the third model
```

```python
plt.figure(figsize=(10,7))
sn.heatmap(cm2,annot =True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')

# Visualize the confusion matrix for the third model as a heatmap
```
